```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np
```

```python
df = pd.read_csv('epa-sea-level.csv')
```

```python
df.tail()
```

```python
year = df['Year']
sea_level = df['CSIRO Adjusted Sea Level']

slope1 = linregress(year, sea_level).slope
intercept1 = linregress(year, sea_level).intercept
```

```python
x1 = np.arange(1880, 2051)
y1 = slope1 * x1 + intercept1
```

```python
year_2000 = df.loc[df['Year'] >= 2000]['Year']
sea_level_2000 = df.loc[df['Year'] >= 2000]['CSIRO Adjusted Sea Level']

slope2 = linregress(year_2000, sea_level_2000).slope
intercept2 = linregress(year_2000, sea_level_2000).intercept
```

```python
x2 = np.arange(2000, 2051)
y2 = slope2 * x2 + intercept2
```

```python
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], linewidth=0.5, edgecolors='black')
plt.plot(x1, y1, linewidth=2, linestyle='dashed', color = 'orange')
plt.plot(x2, y2, linewidth=2, linestyle='dotted', color = 'red')
plt.title('Rise')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.savefig('sea_level.png')
```
